## <span style="color:blue">Minado Masivo</span>

```python
import aiohttp
import asyncio
import json
import pandas as pd
from tqdm.asyncio import tqdm
import nest_asyncio

nest_asyncio.apply()

# URL base
BASE_URL = "https://aiagentsdirectory.com/landscape"

# Función para obtener datos de una URL específica
async def fetch_data(session, url):
    async with session.get(url) as response:
        if response.status == 200:
            return {"url": url, "status": response.status, "data": await response.text()}
        else:
            print(f"Error: {response.status} en {url}")
            return {"url": url, "status": response.status, "data": None}

# Función principal para gestionar las solicitudes concurrentes
async def fetch_all_data(urls):
    async with aiohttp.ClientSession() as session:
        tasks = [fetch_data(session, url) for url in urls]
        results = await tqdm.gather(*tasks, desc="Minando datos...")
        return results

# Procesar los datos extraídos
def process_data(raw_data_list):
    all_agents = []
    validity = []

    for raw_data in raw_data_list:
        if raw_data["data"]:  # Verificar que no sea None
            soup = BeautifulSoup(raw_data["data"], "html.parser")
            script_tag = soup.find('script', {"id": "__NEXT_DATA__"})
            if script_tag:
                directory_json = json.loads(script_tag.text)
                agents = directory_json["props"]["pageProps"]["agent"]
                all_agents.append(agents)

        validity.append({"url": raw_data["url"], "status": raw_data["status"]})
    # Crear DataFrame
    df = pd.DataFrame(all_agents)
    return df, validity

# URLs de ejemplo para scraping
urls_to_scrape = data_agents_filtered["url"].tolist()

# Ejecutar todo el flujo
async def main():
    print("Iniciando scraping concurrente...")
    raw_data = await fetch_all_data(urls_to_scrape)
    dataframe, validity = process_data(raw_data)
    
    print("Scraping completado. DataFrame creado:")
    return dataframe, validity

# Ejecutar
if __name__ == "__main__":
    all_agents_df, validity= asyncio.run(main())
    display(all_agents_df)
    # Convertir el JSON a un diccionario basado en 'link' para acceso rápido
    status_dict = {entry["url"]: entry["status"] for entry in validity}
    # Añadir una nueva columna al DataFrame usando map
    data_agents_filtered.loc[:,"status"] = data_agents_filtered.loc[:,"url"].map(status_dict)
```
