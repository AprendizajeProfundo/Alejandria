```python
# interactive graphs with plotly
# https://plotly.com/python/plotly-express/
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
data = pd.read_parquet("D:/Alejandria/output/models_hf.parquet")
```

```python
# Conteo de modelos por día
data['created_at'] = pd.to_datetime(data['created_at'])
data['created_at'] = data['created_at'].dt.date
data['created_at'] = pd.to_datetime(data['created_at'])
```

```python
conteo_diario = data.groupby(data['created_at'].dt.date).size().reset_index(name="conteo")
conteo_diario.columns = ["created_at", "conteo"]
# Crear la gráfica de serie de tiempo interactiva
fig = px.line(conteo_diario, x="created_at", y="conteo", title='Conteo de modelos por día')
```

```python
fig.show()
```
