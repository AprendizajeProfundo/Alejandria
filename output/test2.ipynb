# Lección de JupyterLab: Análisis de Datos de CHATDEV

En esta lección, aprenderemos a visualizar datos similares a los discutidos en el artículo sobre CHATDEV. Vamos a crear una gráfica de barras para representar la distribución de sugerencias del revisor en el proceso de revisión del código generado por CHATDEV.

## Paso 1: Preparación de Datos

Primero, necesitamos datos de ejemplo. Supongamos que tenemos datos sobre sugerencias del revisor durante la revisión de 100 programas generados por CHATDEV. Los datos están en formato JSON y se parecen a esto:

```json
{
    "methods_not_implemented": 35,
    "modules_not_imported": 20,
    "code_robustness": 5,
    "other_issues": 40
}


````python
import matplotlib.pyplot as plt

# Datos de ejemplo
sugerencias = {
    "methods_not_implemented": 35,
    "modules_not_imported": 20,
    "code_robustness": 5,
    "other_issues": 40
}

# Nombres de las sugerencias para las etiquetas del eje x
sugerencia_nombres = list(sugerencias.keys())

# Cantidad de sugerencias para las alturas de las barras
cantidad_sugerencias = list(sugerencias.values())

# Crear la gráfica de barras
plt.figure(figsize=(8, 6))
plt.bar(sugerencia_nombres, cantidad_sugerencias, color='skyblue')
plt.xlabel('Tipos de Sugerencias del Revisor')
plt.ylabel('Cantidad de Sugerencias')
plt.title('Distribución de Sugerencias del Revisor en CHATDEV')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor claridad
plt.show()
````
